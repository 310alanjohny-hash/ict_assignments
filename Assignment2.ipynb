```python
def calculate_bill(units):
    total_bill = 0
    if units <= 100:
        total_bill = units * 5
    elif units <= 200:
        total_bill = (100 * 5) + ((units - 100) * 7)
    else:
        total_bill = (100 * 5) + (100 * 7) + ((units - 200) * 10)
    return total_bill

# Get units consumed from the user
units_consumed_str = input("Enter units consumed: ")

try:
    units_consumed = int(units_consumed_str)
    if units_consumed < 0:
        print("Units consumed cannot be negative. Please enter a valid number.")
    else:
        bill_amount = calculate_bill(units_consumed)
        print(f"Total Bill Amount: ₹{bill_amount}")
except ValueError:
    print("Invalid input. Please enter a whole number for units consumed.")
```

### Palindrome Checker

This program checks if a given word, phrase, or number is a palindrome (reads the same forwards and backward, ignoring spaces and capitalization).

```python
s = input('Enter num or string to be checked = ')
 if s == s[::-1]:
  print("Given num/string is Palindrome")
  else:
     print("Given num/string is not Palindrome")
```
